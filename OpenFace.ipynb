## OpenCV to Extract Frames From Video

```python
import cv2
import math
import os

def extract_frames(video_path, output_directory):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)

    hour = 0
    minute = 0
    second = 0

    frame_number = 0
    target_frame_rate = 1  
    target_frame_interval = int(fps / target_frame_rate)

    while True:
        success, frame = cap.read()
        if not success:
            break

        if frame_number % target_frame_interval == 0:
            timestamp = frame_number / fps
            hour = math.floor(timestamp / 3600)
            minute = math.floor((timestamp % 3600) / 60)
            second = math.floor(timestamp % 60)

            frame_name = f"{hour:02d}_{minute:02d}_{second:02d}.jpg"
            output_path = os.path.join(output_directory, frame_name)
            cv2.imwrite(output_path, frame)

        frame_number += 1

    cap.release()

video_path = "video_file_path.mp4"
output_directory = "output_directory_path"
extract_frames(video_path, output_directory)
```
